```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro

# =============================
# 1. Data
# =============================

data = np.array([
    7.936, 4.612, 2.407, 4.278, 5.132,
    4.599, 5.224, 2.003, 1.857, 2.696,
    5.259, 7.563, 3.937, 6.908, 5.002,
    6.212, 2.759, 7.172, 6.513, 3.326,
    8.761, 4.502, 6.188, 2.566, 5.515,
    3.785, 3.742, 4.682, 4.346, 5.359,
    3.535, 5.061, 4.629, 5.298, 6.492,
    3.502, 4.266, 3.129, 1.298, 3.454,
    5.289, 6.805, 3.827, 3.912, 2.969,
    4.646, 5.963, 3.829, 4.404, 4.924
])

# =============================
# 2. Statistik deskriptif
# =============================
mean = np.mean(data)
std = np.std(data, ddof=1)        # sample standard deviation
minimum = np.min(data)
maximum = np.max(data)

print("=== Statistik Deskriptif ===")
print(f"Mean            = {mean:.5f}")
print(f"Std Dev (s)     = {std:.5f}")
print(f"Min             = {minimum}")
print(f"Max             = {maximum}\n")

# =============================
# 3. Histogram
# =============================

plt.figure(figsize=(8,5))
plt.hist(data, bins=10, edgecolor='black')
plt.title("Histogram Waktu Transmisi")
plt.xlabel("Waktu (ms)")
plt.ylabel("Frekuensi")
plt.grid(alpha=0.4)
plt.show()

# =============================
# 4. Uji Normalitas Shapiro-Wilk
# =============================

W, p = shapiro(data)

print("=== Uji Shapiro–Wilk ===")
print(f"W statistic = {W}")
print(f"p-value     = {p}\n")

if p > 0.05:
    print("Keputusan: Tidak menolak H0 (data konsisten dengan distribusi Normal).")
else:
    print("Keputusan: Menolak H0 (data tidak berdistribusi Normal).")

# =============================
# 5. Parameter Distribusi Normal
# =============================

print("\n=== Model Distribusi ===")
print(f"T ~ Normal(mean={mean:.2f}, sd={std:.2f})")
```
